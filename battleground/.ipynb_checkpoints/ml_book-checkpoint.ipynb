```python
from IPython.display import clear_output, Image, display, HTML

def strip_consts(graph_def, max_const_size=32):
    """Wyciąga duże wartości stałych z obiektu graph_def."""
    strip_def = tf.GraphDef()
    for n0 in graph_def.node:
        n = strip_def.node.add() 
        n.MergeFrom(n0)
        if n.op == 'Const':
            tensor = n.attr['value'].tensor
            size = len(tensor.tensor_content)
            if size > max_const_size:
                tensor.tensor_content = b"<stripped %d bytes>"%size
    return strip_def

def show_graph(graph_def, max_const_size=32):
    """Wizualizuje graf TensorFlow."""
    if hasattr(graph_def, 'as_graph_def'):
        graph_def = graph_def.as_graph_def()
    strip_def = strip_consts(graph_def, max_const_size=max_const_size)
    code = """
        <script>
          function load() {{
            document.getElementById("{id}").pbtxt = {data};
          }}
        </script>
        <link rel="import" href="https://tensorboard.appspot.com/tf-graph-basic.build.html" onload=load()>
        <div style="height:600px">
          <tf-graph-basic id="{id}"></tf-graph-basic>
        </div>
    """.format(data=repr(str(strip_def)), id='graph'+str(np.random.rand()))

    iframe = """
        <iframe seamless style="width:1200px;height:620px;border:0" srcdoc="{}"></iframe>
    """.format(code.replace('"', '&quot;'))
    display(HTML(iframe))
```

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
import tensorflow as tf

housing = fetch_california_housing()
m, n = housing.data.shape

housing_biassed = np.c_[np.ones((m, 1)), housing.data]
X = tf.constant(housing_biassed, dtype=tf.float64, name='X')
y = tf.constant(housing.target.reshape(-1, 1))
```

```python
# METODA RÓWNANIA NORMALNEGO #

X_T = tf.transpose(X)

# równanie normalne
theta = tf.matmul(tf.matmul(tf.matrix_inverse(tf.matmul(X_T, X)), X_T), y)

sess = tf.InteractiveSession()

print(sess.run(theta))

sess.close()
```

```python
# METODA WSADOWEGO GRADIENTU PROSTEGO #
# konieczna jest normalizacja danych!! #
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
housing_scalled = scaler.fit_transform(housing.data)
housing_biassed_scaled = np.c_[np.ones((m, 1)), housing_scalled]
n_epochs = 1000
learning_rate = 0.01
X = tf.constant(housing_biassed_scaled, dtype=tf.float64, name='X')
y = tf.constant(housing.target.reshape(-1, 1), dtype=tf.float64, name='y')
theta = tf.Variable(tf.random_uniform([ n + 1, 1], -1.0, 1.0, seed=42, dtype=tf.float64), name='theta', dtype=tf.float64)

y_pred = tf.matmul(X, theta, name="prognozy")
error = y_pred - y
mse = tf.reduce_mean(tf.square(error), name='mse')
gradients = 2/m * tf.matmul(tf.transpose(X), error)

training_op = tf.assign(theta, theta - learning_rate * gradients) # training operation

init = tf.global_variables_initializer()

sess = tf.InteractiveSession()

sess.run(init)

for epoch in range(n_epochs):
    if epoch % 100 == 0:
        print(f'Epoka:{epoch}, MSE:{mse.eval()}')
    sess.run(training_op)
best_theta = theta.eval()
print("best theta: ", best_theta)

sess.close()
```

```python
test_ratio = 0.2
test_size = int(m * test_ratio)
X_train = X_moons_biassed[:-test_size]
X_test = X_moons_biassed[-test_size:]
y_train = y_moons_col[:-test_size]
y_test = y_moons_col[-test_size:]
```

```python
def get_random_batch(X_train, y_train, batch_size):
    rnd_indices = np.random.randint(0, len(X_train), batch_size)
    X_batch = X_train[rnd_indices]
    y_batch = y_train[rnd_indices]
    return X_batch, y_batch
```

```python
X_batch, y_batch = get_random_batch(X_train, y_train, 5)
X_batch, y_batch
```

```python
tf.reset_default_graph()

n_inputs = 2 #features set
X = tf.placeholder(tf.float32, shape=(None, n_inputs + 1), name='X')
y = tf.placeholder(tf.float32, shape=(None, 1), name='y')
theta = tf.Variable(tf.random_uniform([n_inputs + 1, 1], -1.0, 1.0, seed=42), name='theta')
logits = tf.matmul(X, theta, name='logits')
# y_probability = 1/(1+tf.exp(-logits))
y_probability = tf.sigmoid(logits)

#
# epsilon = 1e-7
# loss = -tf.reduce_mean(y * tf.log(y_probability + epsilon) + 
#                        (1 - y) * tf.log(1 - y_probability + epsilon))
#
loss = tf.losses.log_loss(y, y_probability)

learning_rate = 0.01
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(loss)

init = tf.global_variables_initializer()

n_epochs = 1000
batch_size = 50
n_batches = int(np.ceil(m/batch_size))

with tf.Session() as sess:
    sess.run(init)
    for epoch in range(n_epochs):
        for batch_index in range(n_batches):
            X_batch, y_batch = get_random_batch(X_train, y_train, batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y:y_batch})
        loss_val = loss.eval(feed_dict={X: X_batch, y:y_batch})
        if epoch %100 == 0:
            print(f"Epoch: {epoch}, Loss: {loss_val}")
    y_prob_val = y_probability.eval(feed_dict={X: X_test, y:y_test})
```

```python
from sklearn.metrics import precision_score, recall_score
y_result = y_prob_val >= 0.5
print(precision_score(y_test, y_result))
print(recall_score(y_test, y_result))
```

```python
# implementacja własna #

import tensorflow as tf
n_inputs = 28*28
n_hidden1 = 300
n_hidden2 = 100
# n_hidden3 = 700
n_outputs = 10


X = tf.placeholder(tf.float32, shape=(None, n_inputs), name='X')
y = tf.placeholder(tf.int64, shape=(None), name='y')

def neuron_layer(X, n_neurons, name, activation=None):
    with tf.name_scope(name):
        n_inputs = int(X.get_shape()[1])
        stddev = 2 / np.sqrt(n_inputs)
        init = tf.truncated_normal((n_inputs, n_neurons), stddev = stddev)
        W = tf.Variable(init, name='kernel')
        b = tf.Variable(tf.zeros([n_neurons]), name='bias')
        Z = tf.matmul(X,W) + b
        if activation is not None:
            return activation(Z)
        else:
            return Z

        
#model sieci
with tf.name_scope('DNN'):
    h1 = neuron_layer(X, 
                n_hidden1, 
                name='hidden_layer_1', 
                activation=tf.nn.relu)
    h2 = neuron_layer(h1, 
                n_hidden2, 
                name='hidden_layer_2',
                activation=tf.nn.relu)
#     h3 = neuron_layer(h2, 
#                 n_hidden3, 
#                 name='hidden_layer_3',
#                 activation=tf.nn.relu)
    logits = neuron_layer(h2,
                n_outputs,
                name='outputs')
    
#funkcja kosztu
with tf.name_scope('loss'):
    x_entropy = tf.nn.sparse_softmax_cross_entropy_with_logits(
        labels=y,logits=logits)
    loss = tf.reduce_mean(x_entropy, name='loss')

    
learning_rate = 0.01
with tf.name_scope('learning'):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
    
with tf.name_scope('score'):
    correct = tf.nn.in_top_k(logits, y, 1)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
    
init = tf.global_variables_initializer()
saver = tf.train.Saver()


```

```python
#faza wykonawcza
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets('/tmp/dane/')
n_epochs = 40
batch_size = 50

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        for iteration in range(mnist.train.num_examples // batch_size):
            X_batch, y_batch = mnist.train.next_batch(batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch})
        acc_train = accuracy.eval(feed_dict={X: X_batch, y: y_batch})
        acc_test = accuracy.eval(feed_dict={X: mnist.validation.images,
                                            y: mnist.validation.labels})
        print(epoch, f'Learning accuracy: {acc_train}, Test set accuracy: {acc_test}')
        save_path = saver.save(sess, './final.ckpt')
```

```python


import tensorflow as tf
import numpy as np

#not saturating activation functions

def leaky_relu(z, name=None, alpha=0.01):
    return tf.maximum(alpha*z, z, name=name)

def elu(z, alpha=1):
    return tf.where(z < 0, alpha * (tf.exp(z) - 1), z)

def selu(z,
         scale=1.0507009873554804934193349852946,
         alpha=1.6732632423543772848170429916717):
    return scale * tf.where(z >= 0.0, z, alpha * tf.nn.elu(z))
#construction phase

tf.reset_default_graph()


#MNIST
n_inputs = 28*28
n_hidden1 = 300
n_hidden2 = 100
n_outputs = 10

X = tf.placeholder(tf.float32, shape=(None, n_inputs), name='X')
y=tf.placeholder(tf.int64, shape=(None), name='y')


#inicjalizacja He
activation_f = selu
# he_init = tf.variance_scaling_initializer() #!!
with tf.name_scope('DNN'):
    hidden1 = tf.layers.dense(X, 
                              n_hidden1, 
                              activation=activation_f, #!!
#                               kernel_initializer=he_init, #!!
                              name='hidden_1')
    hidden2 = tf.layers.dense(hidden1, 
                              n_hidden2, 
                              activation=activation_f, #!!
#                               kernel_initializer=he_init, #!!
                              name='hidden_2')
    logits = tf.layers.dense(hidden2,
                             n_outputs,
                             name='outputs')
    
with tf.name_scope('LOSS'):
    xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y,
                                                        logits=logits)
    loss = tf.reduce_mean(xentropy, name='loss')

learning_rate = 0.01
with tf.name_scope('LEARN'):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
    
with tf.name_scope('SCORE'):
    correct = tf.nn.in_top_k(logits, y, 1)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
    
init = tf.global_variables_initializer()
saver = tf.train.Saver()
tf.summary.FileWriter('logs/train', graph=tf.get_default_graph()).close()


# %load_ext tensorboard.notebook
# %tensorboard --logdir logs/train

```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/dane/")

# # load data
# mnist = tf.keras.datasets.mnist
# (X_train_k, y_train_k), (X_test_k, y_test_k) = mnist.load_data()
# X_train = (X_train_k / 255)
# X_test = (X_test_k / 255)
# y_train = y_train_k.astype('int')
# y_test = y_test_k.astype('int')

n_epochs = 55
batch_size = 50

means = mnist.train.images.mean(axis=0, keepdims=True)
stds = mnist.train.images.std(axis=0, keepdims=True) + 1e-10

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        for iteration in range(mnist.train.num_examples // batch_size):
            X_batch, y_batch = mnist.train.next_batch(batch_size)
            X_batch_scaled = (X_batch - means) / stds
            sess.run(training_op, feed_dict={X: X_batch_scaled, y: y_batch})
        if epoch % 5 == 0:
            acc_train = accuracy.eval(feed_dict={X: X_batch_scaled, y: y_batch})
            X_val_scaled = (mnist.validation.images - means) / stds
            acc_test = accuracy.eval(feed_dict={X: X_val_scaled, y: mnist.validation.labels})
            print(epoch, f" batch loss: {acc_train:.5f}, validation loss: {acc_test:.5f}")
    saver.save(sess, './final.ckpt')
```

## Clipping gradients

```python
tf.reset_default_graph()


n_inputs = 28*28
n_hidden1 = 300
n_hidden2 = 50
n_hidden3 = 50
n_hidden4 = 50
n_hidden5 = 50
n_outputs = 10


X = tf.placeholder(tf.float32, shape=(None, n_inputs), name='X')
y = tf.placeholder(tf.int64, shape=(None), name='y')

with tf.name_scope('DNN'):
    h1 = tf.layers.dense(X,  n_hidden1, activation=tf.nn.relu, name='hidden_1')
    h2 = tf.layers.dense(h1, n_hidden2, activation=tf.nn.relu, name='hidden_2')
    h3 = tf.layers.dense(h2, n_hidden3, activation=tf.nn.relu, name='hidden_3')
    h4 = tf.layers.dense(h3, n_hidden4, activation=tf.nn.relu, name='hidden_4')
    h5 = tf.layers.dense(h4, n_hidden5, activation=tf.nn.relu, name='hidden_5')
    logits =tf.layers.dense(h5, n_outputs, name='outputs')

with tf.name_scope('LOSS'):
    xentropy =tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, 
                                                             logits=logits)
    loss = tf.reduce_mean(xentropy, name='loss')

learning_rate = 0.01
treshold = 1.0

optimizer = tf.train.GradientDescentOptimizer(learning_rate)

##Clipping operation############
grads_and_vars = optimizer.compute_gradients(loss)
capped_gvs = [(tf.clip_by_value(grad, -treshold, treshold), var) for grad, var in grads_and_vars]
training_op = optimizer.apply_gradients(capped_gvs)
################################

with tf.name_scope('SCORE'):
    correct = tf.nn.in_top_k(logits, y, 1)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32), name='accuracy')
    
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
n_epochs = 40
batch_size = 50

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        for iteration in range(mnist.train.num_examples // batch_size):
            X_batch, y_batch = mnist.train.next_batch(batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch})
        acc_val = accuracy.eval(feed_dict={X: mnist.test.images,
                                          y: mnist.test.labels})
        print(f"{epoch} => ACCURACY: {acc_val}")
        
    saver.save(sess, './final.ckpt')
```
