```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib qt
```

```python

datasets_path = './datasets'
dataset_name = 'master.csv'

full_path = os.path.join(datasets_path, dataset_name)

suicide_df = pd.read_csv(full_path)

suicide_df.head(5)

label = suicide_df['suicides_no']
features = suicide_df.drop(['suicides/100k pop', 'suicides_no'], axis=1)
```

```python
plt.scatter(features.year,label, label='suicide_no')
```

```python
pol_idxs = suicide_df.country == 'Poland'

suicide_df_pol = suicide_df[pol_idxs]

# pol_gen_x_idxs = suicide_df_pol.generation == 'Generation X'

# suicide_df_pol = suicide_df_pol[pol_gen_x_idxs]

# suicide_df_pol.count()
suicide_df_pol.describe()
```

```python
plot_no = 1

for age in set(suicide_df_pol.age):
    suicide_df_pol_filtered = suicide_df_pol[suicide_df_pol.age == age]
    ax = plt.subplot(3,2,plot_no)
    plt.title(f'suicides in age group: {age}')
    for generation in set(suicide_df_pol_filtered.generation):
        idxs = suicide_df_pol_filtered.generation == generation
        sparse_df = suicide_df_pol_filtered[idxs]
        x = sparse_df.year.unique()
        y_male = sparse_df[sparse_df.sex == 'male'].suicides_no
        y_female = sparse_df[sparse_df.sex == 'female'].suicides_no
        print(len(x), len(y_male), len(y_female))
        ax1 = plt.scatter(x, y_male, label=f'{generation} - male')
        ax2 = plt.scatter(x, y_female, label=f'{generation} - female')
    plt.legend()
    plt.grid()
    plot_no += 1
```
